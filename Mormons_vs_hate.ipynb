# MORMONS VS HATE

Fast ckeck about correlation between mormons [stakes](https://en.wikipedia.org/wiki/Stake_%28Latter_Day_Saints%29) numbers and [hate groups](https://www.splcenter.org/fighting-hate/intelligence-report/2017/active-hate-groups-2016) numbers

_RESULT: No Correlation_


```python
#importing libraries 
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading data
df = pd.read_csv('./mormon_vs_hate.csv')
#filling NaN values with 0
df.fillna(0)

# DATA COLUMS ARE
# NAME10 --> State name
# DP0010001 --> Population 
# PNTCNT --> number of hate groups
# MORCNT --> number of mormons stakes
```

```python
#plotting data
plt.scatter(df['PNTCNT'], df['MORCNT'])
plt.xlabel('Hate groups')
plt.ylabel('Mormon stake')
```

```python
#calculating correlation

df['PNTCNT'].corr(df['MORCNT'])
```

Correlation between hate group numbers by state and mormons stake by state is very close to zero. ***This means that the two variables do not share any particular correlation***.
